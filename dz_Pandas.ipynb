Задание 1
Скачайте с сайта grouplens.org...movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
movies_ratings = ratings.merge(movies, on='movieId', how='left')
movies_ratings.head()
```

```python
movies_ratings = movies_ratings[['movieId','rating','title']] 
movies_ratings.head()
```

```python
max_rating = ratings['rating'].max()
filter_max = movies_ratings[(movies_ratings ['rating'] == max_rating)] 
filter_max.head()
```

```python
movie_rating_5 = filter_max.groupby(['movieId' , 'title'])['movieId'].count()
movie_rating_5.sort_values(ascending = False).head(1)
```
