```python
Выводы: 
Значения параметра Surgery min=1, max=2. Выбросов скорее всего нет. Так как данный параметр должен принимать значение 1 и 2. Большая часть прооперированных лошадей являлась взрослыми, старше 6 месяцев.   
Значения параметра Age по условиям может быть только 1 или 2. Инные значения, max=9, скорее всего были введены ошибочно и являются выбросами
```

```python
# Найдем выбросы в числовых столбцах : rectal temperature и pulse
```

```python
# Определим выбросы в rectal temperature используя межквартильный размах.
IQR_RT = new_table['rectal temperature'].quantile(0.75) - df['rectal temperature'].quantile(0.25)
lower_bound = new_table['rectal temperature'].quantile(0.25) - 1.5 * IQR_RT
upper_bound = new_table['rectal temperature'].quantile(0.75) + 1.5 * IQR_RT
print(lower_bound, upper_bound)
```

```python
new_table[(new_table['rectal temperature'] < lower_bound) | 
                         (new_table['rectal temperature'] > upper_bound)]
```

```python
# Выводы:значения rectal temperature выходят за нижнюю и верхнюю границу нормы. Возможно являются выбросами, либо же такая температура имела место быть
```

```python
# Определим выбросы в pulse используя межквартильный размах.
IQR_pulse = new_table['pulse'].quantile(0.75) - df['pulse'].quantile(0.25)
lower_bound = new_table['pulse'].quantile(0.25) - 1.5 * IQR_pulse
upper_bound = new_table['pulse'].quantile(0.75) + 1.5 * IQR_pulse
print(lower_bound, upper_bound)
```

```python
new_table[(new_table['pulse'] < lower_bound) | 
                         (new_table['pulse'] > upper_bound)]
```

```python
# Выводы: выбросы в значениях пульсах коснулись всех лошадей , у которых в колонке Age указано значение 9, что скорее всего является выбросом. Соответственно, можем предположить что данные показатели пульса так же были введены ошибочно. 
```

Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.


```python
# Рассчитаем количество пропусков
new_table.isna().sum()
```

```python
# Заполним пропуски в числовых значениях медианой, в категориальных значениях модой. 

new_table['surgery?'].fillna(new_table['surgery?'].median(), inplace=True)
new_table['rectal temperature'].fillna(new_table['rectal temperature'].median(), inplace=True)
new_table['pulse'].fillna(new_table['pulse'].median(), inplace=True)
new_table['temperature of extremities'].fillna(new_table['temperature of extremities'].mode()[0], inplace=True)
new_table['peripheral pulse'].fillna(new_table['peripheral pulse'].mode()[0], inplace=True)                   
new_table['pain'].fillna(new_table['pain'].mode()[0], inplace=True)
```

```python
# сформируем датафрейм, в котором пропуски отсутствуют
new_table.isna().sum()
```
